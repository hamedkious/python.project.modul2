```python
from scipy.io import arff
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
```

```python
file_name = "C:\\Users\\hamed.ghaseminezhad\\Downloads\\project_data_set.arff"
data = arff.loadarff(file_name)
df = pd.DataFrame(data[0])
```

```python
df.classe = df.classe.apply(lambda cell: cell.decode())
df.head()
```

```python
df.shape
```

```python
X = df.drop(columns=['classe'])
y = df.classe
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, train_size=.6)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model = DecisionTreeClassifier()
model.fit(X_train, y_train);
```

```python
predicted_y_train = model.predict(X_train)
predicted_y_test = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score

print(accuracy_score(y_train, predicted_y_train))
print(accuracy_score(y_test, predicted_y_test))
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(200, 150))
plot_tree(decision_tree=model, filled=True, feature_names=df.columns[:-1], class_names=df.classe.unique());
```

```python
import math
def euclidean(x,y):
    dist=0
    for i in range(len(x)):
        dist+=(x[i]-y[i])**2
    return math.sqrt(dist)
```
